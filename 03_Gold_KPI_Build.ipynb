## 1. KPI

```python
%sql
-- NOTEBOOK 03: GOLD KPI BUILD
-- PURPOSE: Create business-ready KPI table from completed Silver layer

CREATE OR REPLACE TABLE claims_data_analysis.claims_db.gold_kpi_summary_2026 AS

SELECT
    COUNT(*) AS total_claim_rows,

    SUM(AMT_HKD) AS total_claim_amount_hkd,
    AVG(AMT_HKD) AS avg_claim_amount_hkd,
    MIN(AMT_HKD) AS min_claim_amount_hkd,
    MAX(AMT_HKD) AS max_claim_amount_hkd,

    COUNT(DISTINCT BENEFIT_CD) AS distinct_benefit_codes,
    COUNT(DISTINCT SUN_CD) AS distinct_sun_codes,
    COUNT(DISTINCT CHANNEL) AS distinct_channels,
    COUNT(DISTINCT GEOGRAPHY) AS distinct_geographies,

    COUNT(DISTINCT benefit_level_1) AS distinct_benefit_categories,
    COUNT(DISTINCT sun_claim_type) AS distinct_sun_claim_types,
    COUNT(DISTINCT channel_name) AS distinct_channel_names,
    COUNT(DISTINCT geography_country) AS distinct_countries

FROM claims_data_analysis.claims_db.claims_silver_2026;
```

```python
%sql
SELECT *
FROM claims_data_analysis.claims_db.gold_kpi_summary_2026;
```

## 2. Benefit Claims Summary

```python
%sql
-- GOLD TABLE 2: BENEFIT CLAIMS SUMMARY
-- PURPOSE: Claims performance by benefit category

CREATE OR REPLACE TABLE claims_data_analysis.claims_db.gold_benefit_summary_2026 AS

SELECT
    benefit_level_1,
    benefit_level_2,
    benefit_description,

    COUNT(*) AS claim_count,
    SUM(AMT_HKD) AS total_claim_amount_hkd,
    AVG(AMT_HKD) AS avg_claim_amount_hkd,
    MIN(AMT_HKD) AS min_claim_amount_hkd,
    MAX(AMT_HKD) AS max_claim_amount_hkd

FROM claims_data_analysis.claims_db.claims_silver_2026

GROUP BY
    benefit_level_1,
    benefit_level_2,
    benefit_description

ORDER BY total_claim_amount_hkd DESC;
```

## 3. Channel Summary

```python
%sql
-- GOLD TABLE 3: CHANNEL CLAIMS SUMMARY
-- PURPOSE: Claims performance by instruction/channel source

CREATE OR REPLACE TABLE claims_data_analysis.claims_db.gold_channel_summary_2026 AS

SELECT
    CHANNEL AS channel_code,
    channel_name,
    channel_description,

    COUNT(*) AS claim_count,
    SUM(AMT_HKD) AS total_claim_amount_hkd,
    AVG(AMT_HKD) AS avg_claim_amount_hkd,
    MIN(AMT_HKD) AS min_claim_amount_hkd,
    MAX(AMT_HKD) AS max_claim_amount_hkd

FROM claims_data_analysis.claims_db.claims_silver_2026

GROUP BY
    CHANNEL,
    channel_name,
    channel_description

ORDER BY total_claim_amount_hkd DESC;
```

```python
%sql
SELECT *
FROM claims_data_analysis.claims_db.gold_channel_summary_2026
LIMIT 50;
```

## 4. Claims By Geography

## 6. Monthly Claims Summary

```python
%sql
-- GOLD TABLE 6: MONTHLY CLAIMS SUMMARY
-- PURPOSE: Analyze claims by accounting/reporting month

CREATE OR REPLACE TABLE claims_data_analysis.claims_db.gold_monthly_summary_2026 AS

SELECT
    ACCT_PERIOD,

    COUNT(*) AS claim_count,
    SUM(AMT_HKD) AS total_claim_amount_hkd,
    AVG(AMT_HKD) AS avg_claim_amount_hkd,
    MIN(AMT_HKD) AS min_claim_amount_hkd,
    MAX(AMT_HKD) AS max_claim_amount_hkd,

    COUNT(DISTINCT POLICY_NO) AS distinct_policies,
    COUNT(DISTINCT BENEFIT_CD) AS distinct_benefit_codes,
    COUNT(DISTINCT SUN_CD) AS distinct_sun_codes,
    COUNT(DISTINCT CHANNEL) AS distinct_channels,
    COUNT(DISTINCT GEOGRAPHY) AS distinct_geographies

FROM claims_data_analysis.claims_db.claims_silver_2026

GROUP BY
    ACCT_PERIOD

ORDER BY
    ACCT_PERIOD;
```

```python
%sql
SELECT *
FROM claims_data_analysis.claims_db.gold_monthly_summary_2026
ORDER BY ACCT_PERIOD;
```
